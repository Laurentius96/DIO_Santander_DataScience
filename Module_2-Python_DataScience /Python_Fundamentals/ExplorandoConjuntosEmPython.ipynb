## üöÄ Aula 1: Explorando Conjuntos em Python

```python
# Se√ß√£o para suas anota√ß√µes sobre conjuntos em Python

"""
O que s√£o conjuntos (sets) em Python:
- Cole√ß√µes n√£o ordenadas de elementos √∫nicos
- N√£o permitem elementos duplicados
- S√£o mut√°veis (podem ser alterados ap√≥s a cria√ß√£o)
- S√£o delimitados por chaves {} ou criados com a fun√ß√£o set()
- N√£o s√£o index√°veis (n√£o podemos acessar elementos por √≠ndice)
- Implementam opera√ß√µes matem√°ticas de conjuntos (uni√£o, interse√ß√£o, etc.)
"""

# Exemplo b√°sico de conjunto
meu_primeiro_conjunto = {1, 2, 3, 4, 5}
print(meu_primeiro_conjunto)

# Conjunto com elementos duplicados (os duplicados s√£o automaticamente removidos)
conjunto_com_duplicatas = {1, 2, 2, 3, 3, 3, 4, 5, 5}
print(conjunto_com_duplicatas)  # {1, 2, 3, 4, 5}

# Verificando o tipo
print(type(meu_primeiro_conjunto))  # <class 'set'>

# Tamanho do conjunto
print(len(meu_primeiro_conjunto))  # 5

# Compara√ß√£o com listas (que permitem duplicatas e s√£o ordenadas)
minha_lista = [1, 2, 2, 3, 3, 3, 4, 5, 5]
print(f"Lista: {minha_lista}")
print(f"Lista convertida para conjunto: {set(minha_lista)}")
```

## üîç Cria√ß√£o e Manipula√ß√£o de Conjuntos

```python
# Criando conjuntos
conjunto_vazio = set()  # Conjunto vazio (n√£o podemos usar {} pois isso cria um dicion√°rio vazio)
print(conjunto_vazio)

# Criando conjunto a partir de uma string
conjunto_string = set("Python")
print(conjunto_string)  # {'P', 'y', 't', 'h', 'o', 'n'}

# Criando conjunto a partir de uma lista
conjunto_lista = set([1, 2, 3, 2, 1])
print(conjunto_lista)  # {1, 2, 3}

# Criando conjunto a partir de uma tupla
conjunto_tupla = set((1, 2, 3, 2, 1))
print(conjunto_tupla)  # {1, 2, 3}

# Adicionando elementos
frutas = {"ma√ß√£", "banana", "laranja"}
print(frutas)

frutas.add("uva")
print(frutas)

# Tentando adicionar um elemento que j√° existe (n√£o causa erro, mas n√£o modifica o conjunto)
frutas.add("ma√ß√£")
print(frutas)  # A ma√ß√£ n√£o √© adicionada novamente

# Adicionando m√∫ltiplos elementos
frutas.update(["morango", "abacaxi", "kiwi"])
print(frutas)

# Tamb√©m podemos usar update com outros conjuntos, tuplas, strings, etc.
frutas.update(("p√™ssego", "melancia"))
print(frutas)

# Removendo elementos
# remove() - remove um elemento, mas gera erro se ele n√£o existir
frutas.remove("banana")
print(frutas)

try:
    frutas.remove("banana")  # Vai gerar erro
except KeyError as e:
    print(f"Erro: {e}")

# discard() - remove um elemento, mas n√£o gera erro se ele n√£o existir
frutas.discard("laranja")
print(frutas)

frutas.discard("laranja")  # N√£o gera erro
print(frutas)

# pop() - remove e retorna um elemento arbitr√°rio
elemento = frutas.pop()
print(f"Elemento removido: {elemento}")
print(frutas)

# clear() - remove todos os elementos
frutas.clear()
print(frutas)  # set()
```

## üõ†Ô∏è Opera√ß√µes com Conjuntos

```python
# Definindo alguns conjuntos para exemplos
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
C = {1, 2}

# Verificando se um elemento pertence ao conjunto
print(1 in A)  # True
print(6 in A)  # False

# Verificando se um conjunto √© subconjunto de outro
print(C.issubset(A))  # True (C ‚äÜ A)
print(A.issubset(B))  # False

# Verificando se um conjunto √© superconjunto de outro
print(A.issuperset(C))  # True (A ‚äá C)
print(B.issuperset(A))  # False

# Verificando se dois conjuntos s√£o disjuntos (n√£o t√™m elementos em comum)
print(A.isdisjoint(B))  # False (A e B t√™m elementos em comum: 4 e 5)
D = {10, 11, 12}
print(A.isdisjoint(D))  # True (A e D n√£o t√™m elementos em comum)

# Uni√£o de conjuntos (A ‚à™ B)
uniao = A.union(B)
print(f"A ‚à™ B = {uniao}")

# Uni√£o usando o operador |
uniao_alt = A | B
print(f"A | B = {uniao_alt}")

# Interse√ß√£o de conjuntos (A ‚à© B)
intersecao = A.intersection(B)
print(f"A ‚à© B = {intersecao}")

# Interse√ß√£o usando o operador &
intersecao_alt = A & B
print(f"A & B = {intersecao_alt}")

# Diferen√ßa de conjuntos (A - B)
diferenca = A.difference(B)
print(f"A - B = {diferenca}")

# Diferen√ßa usando o operador -
diferenca_alt = A - B
print(f"A - B = {diferenca_alt}")

# Diferen√ßa sim√©trica (A ‚ñ≥ B) ou (A ‚äï B) - elementos que est√£o em A ou B, mas n√£o em ambos
dif_simetrica = A.symmetric_difference(B)
print(f"A ‚ñ≥ B = {dif_simetrica}")

# Diferen√ßa sim√©trica usando o operador ^
dif_simetrica_alt = A ^ B
print(f"A ^ B = {dif_simetrica_alt}")
```

## üîÑ Opera√ß√µes de Conjuntos com Atribui√ß√£o

```python
# Estas opera√ß√µes modificam o conjunto original

# Uni√£o com atribui√ß√£o (A = A ‚à™ B)
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
A.update(B)  # ou A |= B
print(f"A ap√≥s A.update(B): {A}")

# Interse√ß√£o com atribui√ß√£o (A = A ‚à© B)
A = {1, 2, 3, 4, 5}
A.intersection_update(B)  # ou A &= B
print(f"A ap√≥s A.intersection_update(B): {A}")

# Diferen√ßa com atribui√ß√£o (A = A - B)
A = {1, 2, 3, 4, 5}
A.difference_update(B)  # ou A -= B
print(f"A ap√≥s A.difference_update(B): {A}")

# Diferen√ßa sim√©trica com atribui√ß√£o (A = A ‚ñ≥ B)
A = {1, 2, 3, 4, 5}
A.symmetric_difference_update(B)  # ou A ^= B
print(f"A ap√≥s A.symmetric_difference_update(B): {A}")
```

## üí° Conjuntos Imut√°veis (frozenset)

```python
# Frozenset √© uma vers√£o imut√°vel de conjuntos
conjunto_normal = {1, 2, 3}
conjunto_imutavel = frozenset([1, 2, 3])

print(type(conjunto_normal))    # <class 'set'>
print(type(conjunto_imutavel))  # <class 'frozenset'>

# Opera√ß√µes que funcionam em frozenset
print(2 in conjunto_imutavel)  # True
print(len(conjunto_imutavel))  # 3

# Uni√£o, interse√ß√£o, etc. (retornam novos frozensets)
outro_conjunto = frozenset([3, 4, 5])
uniao = conjunto_imutavel.union(outro_conjunto)
print(uniao)  # frozenset({1, 2, 3, 4, 5})

# Opera√ß√µes que n√£o funcionam em frozenset (pois √© imut√°vel)
try:
    conjunto_imutavel.add(4)
except AttributeError as e:
    print(f"Erro: {e}")

try:
    conjunto_imutavel.remove(1)
except AttributeError as e:
    print(f"Erro: {e}")

# Uso de frozenset como chave de dicion√°rio (conjuntos normais n√£o podem ser usados como chaves)
dicionario = {frozenset([1, 2]): "conjunto imut√°vel"}
print(dicionario[frozenset([1, 2])])  # "conjunto imut√°vel"

try:
    dicionario[{1, 2}] = "isso vai dar erro"
except TypeError as e:
    print(f"Erro: {e}")  # Erro: unhashable type: 'set'
```

## üîç Casos de Uso para Conjuntos

```python
# 1. Remo√ß√£o de duplicatas de uma lista
lista_com_duplicatas = [1, 2, 3, 1, 2, 4, 5, 4, 6]
lista_sem_duplicatas = list(set(lista_com_duplicatas))
print(f"Lista original: {lista_com_duplicatas}")
print(f"Lista sem duplicatas: {lista_sem_duplicatas}")

# 2. Verifica√ß√£o de pertin√™ncia (membership testing) - mais eficiente que em listas
# Criando um conjunto e uma lista grandes
import time

tamanho = 10000
conjunto_grande = set(range(tamanho))
lista_grande = list(range(tamanho))

# Medindo tempo para verificar pertin√™ncia em um conjunto
inicio = time.time()
elemento_in_conjunto = 9999 in conjunto_grande
fim = time.time()
tempo_conjunto = fim - inicio
print(f"Tempo para verificar em conjunto: {tempo_conjunto:.10f} segundos")

# Medindo tempo para verificar pertin√™ncia em uma lista
inicio = time.time()
elemento_in_lista = 9999 in lista_grande
fim = time.time()
tempo_lista = fim - inicio
print(f"Tempo para verificar em lista: {tempo_lista:.10f} segundos")
print(f"O conjunto √© {tempo_lista/tempo_conjunto:.1f}x mais r√°pido")

# 3. Encontrando elementos comuns entre cole√ß√µes
lista1 = [1, 2, 3, 4, 5]
lista2 = [4, 5, 6, 7, 8]
elementos_comuns = set(lista1).intersection(set(lista2))
print(f"Elementos comuns: {elementos_comuns}")

# 4. Encontrando elementos √∫nicos entre cole√ß√µes
elementos_unicos = set(lista1).symmetric_difference(set(lista2))
print(f"Elementos √∫nicos: {elementos_unicos}")

# 5. Verificando se uma cole√ß√£o cont√©m todos os elementos de outra
print(set([1, 2]).issubset(set([1, 2, 3, 4])))  # True
print(set([1, 2, 5]).issubset(set([1, 2, 3, 4])))  # False
```

## üß© Desafio: An√°lise de Texto com Conjuntos

```python
# Desafio: Crie um programa que analise duas strings e encontre:
# 1. Caracteres comuns entre as duas strings
# 2. Caracteres que aparecem apenas na primeira string
# 3. Caracteres que aparecem apenas na segunda string
# 4. Todos os caracteres √∫nicos que aparecem em ambas as strings

def analisar_texto(texto1, texto2):
    # Converta os textos para conjuntos de caracteres
    conjunto1 = set(texto1.lower())
    conjunto2 = set(texto2.lower())
    
    # Caracteres comuns
    comuns = conjunto1.intersection(conjunto2)
    
    # Caracteres exclusivos da primeira string
    exclusivos1 = conjunto1.difference(conjunto2)
    
    # Caracteres exclusivos da segunda string
    exclusivos2 = conjunto2.difference(conjunto1)
    
    # Todos os caracteres √∫nicos
    todos_unicos = conjunto1.union(conjunto2)
    
    return {
        'comuns': comuns,
        'exclusivos1': exclusivos1,
        'exclusivos2': exclusivos2,
        'todos_unicos': todos_unicos
    }

# Exemplo de uso
texto1 = "Python √© uma linguagem de programa√ß√£o"
texto2 = "Java tamb√©m √© uma linguagem de programa√ß√£o"

resultado = analisar_texto(texto1, texto2)

print(f"Caracteres comuns: {sorted(resultado['comuns'])}")
print(f"Exclusivos de texto1: {sorted(resultado['exclusivos1'])}")
print(f"Exclusivos de texto2: {sorted(resultado['exclusivos2'])}")
print(f"Todos os caracteres √∫nicos: {sorted(resultado['todos_unicos'])}")
```

## üìö Conjuntos vs Listas vs Dicion√°rios: Quando Usar Cada Um

```python
"""
Quando usar conjuntos (sets):
1. Quando precisar armazenar elementos √∫nicos
2. Quando a ordem dos elementos n√£o importar
3. Para opera√ß√µes matem√°ticas de conjuntos (uni√£o, interse√ß√£o, etc.)
4. Para verifica√ß√£o r√°pida de pertin√™ncia (in)
5. Para remover duplicatas de uma sequ√™ncia

Quando usar listas:
1. Quando a ordem dos elementos importar
2. Quando precisar de elementos duplicados
3. Quando precisar acessar elementos por √≠ndice
4. Quando precisar modificar elementos espec√≠ficos

Quando usar dicion√°rios:
1. Quando precisar de pares chave-valor
2. Para acesso r√°pido a valores por chave
3. Quando precisar mapear valores a chaves √∫nicas

Compara√ß√£o de desempenho:
"""

import time
import random

# Criando estruturas de dados com 100.000 elementos
numeros = list(range(100000))
random.shuffle(numeros)  # Embaralha a lista

# Criando c√≥pias dos primeiros 10.000 elementos para teste
teste = numeros[:10000]

# Convertendo para diferentes estruturas
lista = numeros.copy()
conjunto = set(numeros)
dicionario = {num: num for num in numeros}

# Teste de pertin√™ncia (verificar se um elemento est√° na estrutura)
print("Teste de pertin√™ncia (verificar se 10.000 elementos est√£o na estrutura):")

# Lista
inicio = time.time()
for num in teste:
    _ = num in lista
fim = time.time()
print(f"  Lista: {fim - inicio:.6f} segundos")

# Conjunto
inicio = time.time()
for num in teste:
    _ = num in conjunto
fim = time.time()
print(f"  Conjunto: {fim - inicio:.6f} segundos")

# Dicion√°rio
inicio = time.time()
for num in teste:
    _ = num in dicionario
fim = time.time()
print(f"  Dicion√°rio: {fim - inicio:.6f} segundos")

# Teste de adi√ß√£o de elementos
print("\nTeste de adi√ß√£o de 10.000 elementos:")

# Lista
inicio = time.time()
lista_teste = []
for i in range(10000):
    lista_teste.append(i)
fim = time.time()
print(f"  Lista (append): {fim - inicio:.6f} segundos")

# Conjunto
inicio = time.time()
conjunto_teste = set()
for i in range(10000):
    conjunto_teste.add(i)
fim = time.time()
print(f"  Conjunto (add): {fim - inicio:.6f} segundos")

# Dicion√°rio
inicio = time.time()
dict_teste = {}
for i in range(10000):
    dict_teste[i] = i
fim = time.time()
print(f"  Dicion√°rio: {fim - inicio:.6f} segundos")

# Tamanho em mem√≥ria
import sys
print("\nTamanho em mem√≥ria:")
print(f"  Lista: {sys.getsizeof(lista)} bytes")
print(f"  Conjunto: {sys.getsizeof(conjunto)} bytes")
print(f"  Dicion√°rio: {sys.getsizeof(dicionario)} bytes")
```
